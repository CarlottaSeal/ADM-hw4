### (a)

First differentiate \(g(x)\) twice:

$$
g'(x)=
\begin{cases}
\frac{2}{T}\left(\frac{1}{U}-\frac{1}{L}\right), & 0\le x\le L, \\
\frac{2}{T}\left(\frac{1}{U}-\frac{1}{x}\right), & L<x<U, \\
0, & x\ge U,
\end{cases}
$$

$$
g''(x)=
\begin{cases}
\frac{2}{T x^2}, & L<x<U, \\
0, & \text{otherwise}.
\end{cases}
$$

These may be further simplified as

$$
g'(x)=\frac{2}{T x}
\left[
\left(\frac{L-x}{L}\right)^+
-
\left(\frac{U-x}{U}\right)^+
\right],
\qquad
g''(x)=\frac{2}{T x^2}1_{(L,U)}(x).
$$

Apply Itô’s Lemma to $g(S_t)$:

$$
d(g(S_t))
=
g'(S_t)\,dS_t
+\frac12 g''(S_t)(dS_t)^2
=
g'(S_t)S_t\mu\,dt
+g'(S_t)S_t\sqrt{v_t}\,dW_t
+\frac12 g''(S_t)S_t^2v_t\,dt.
$$

Integrating and rearranging:

$$
\frac1T\int_0^T 1_{(L,U)}(S_t)v_t\,dt
=
g(S_T)-g(S_0)
-\int_0^T g'(S_t)S_t\mu\,dt
-\int_0^T g'(S_t)S_t\sqrt{v_t}\,dW_t.
$$

Thus,

$$
X = g(S_T)-g(S_0)
-\int_0^T g'(S_t)S_t\mu\,dt
-\int_0^T g'(S_t)S_t\sqrt{v_t}\,dW_t
-H.
$$


### (b)

Using similar method in part (a), if we expand X under the risk-neutral Q measure,$W^Q$ is a Brownian motion under Q, then

$$
X
= g(S_T) - g(S_0)
- \int_0^T g'(S_t)S_t r\,dt
- \int_0^T g'(S_t)S_t\sqrt{v_t}\,dW^Q_t
- H. \tag{1}
$$

Each term in (1) may be replicated as follows.

---

#### **1. Static replication of**  \(g(S_T)\)

Applying the **Carr–Madan Theorem**,

$$
g(S_T)
= g(a) + g'(a)(S_T-a)
+ \int_{0}^{a} g''(K)(K-S_T)^+\,dK
+ \int_{a}^{\infty} g''(K)(S_T-K)^+\,dK.
$$

for any fixed $a \geq 0$. This results in
* a riskfree asset position worth $e^{-rT}(g(a)-g'(a)a)$
* $g'(a)$ shares of the risky asset, and
* an infinite strip of European puts and calls expiring at time $T$ with amount $g′′(K)$ held in options with strike price K.

---

#### **2. Static replication of the constant term**

The term  
$$
-g(S_0) - H
$$
is constant and therefore replicated by a static position in the **risk-free asset**
with value
$$
-e^{-rT}[g(S_0) + H].
$$

---

#### **3. Replication of the drift term**
$$
-\int_0^T g'(S_t)S_t r \, dt.
$$

Since from part (a),

$$
g'(S_t)S_t
= \frac{2}{T}
\left[
\frac{1}{U}(U-S_t)^+ - \frac{1}{L}(L-S_t)^+
\right],
$$

it follows that

$$
-\int_0^T g'(S_t)S_t r \, dt
= \int_0^T \frac{2r}{T}
\left[
\frac{1}{U}(U-S_t)^+ - \frac{1}{L}(L-S_t)^+
\right] dt.
$$

This can be replicated by **two infinite strips of European puts** with expiration  
times \(t \in [0,T]\):

- **Long put** with strike $U$, amount $\frac{2r}{T U} e^{-r(T-t)} $,  
- **Short put** with strike $L$, amount $ -\frac{2r}{T L} e^{-r(T-t)}$.

As each put expires, its proceeds are invested in the money market account until time $T$.

This replicates the drift term in (1).

---

#### **4. Dynamic replication of**
$$
-\int_0^T g'(S_t)S_t\sqrt{v_t}\,dW^Q_t.
$$

Define

$$
V_t = -e^{-r(T-t)}\int_0^t g'(S_s)S_s \sqrt{v_s}\, dW^Q_s.
$$

Applying Itô’s Lemma and substituting $S_t\sqrt{v_t}\,dW^Q_t = -S_t r\,dt + dS_t$, we get

$$dV_t = rV_t\,dt - e^{-r(T-t)}g'(S_t)S_t\sqrt{v_t}\,dW_t$$
$$= \left(V_t + e^{-r(T-t)}g'(S_t)S_t\right)r\,dt - e^{-r(T-t)}g'(S_t)\,dS_t$$


Thus, $V_t$ is a **self-financing dynamic strategy** that holds

- $-e^{-\mu(T-t)}g'(S_t)$ shares of the stock,
- the remainder $(V_t + e^{-r(T-t)}g'(S_t)S_t)$ in the money market account.

At maturity,

$$
V_T = -\int_0^T g'(S_t) S_t \sqrt{v_t}\, dW_t,
$$

replicating the fourth term in (1).

---

### Combining all of the above positions with r=0 exactly replicates the payoff of the corridor variance swap.



### (c) Break-even corridor variance strike

Let  
$$P(T, K) = \mathbb{E}^{\mathbb{Q}}\left[e^{-rT}(K - S_T)^+ \right], 
\qquad
C(T, K) = \mathbb{E}^{\mathbb{Q}}\left[e^{-rT}(S_T - K)^+ \right]$$  
denote European put and call prices with maturity \(T\) and strike \(K\).  

Taking expectations of the relation from part (b),  
$$
\mathbb{E}^{\mathbb{Q}}\left[ \frac{1}{T} \int_0^T 1_{(L,U)}(S_t)v_t dt \right]
= \mathbb{E}^{\mathbb{Q}}[g(S_T)] - g(S_0)
- \mathbb{E}^{\mathbb{Q}}\left[ \int_0^T g'(S_t) S_t r dt \right].
$$

---

### Applying Carr–Madan to $ g(S_T)$ 

Using the Carr–Madan Theorem,
$$
g(S_T) =
g(a) + g'(a)(S_T - a)
+ \int_0^a g''(K)(K-S_T)^+ \, dK
+ \int_a^\infty g''(K)(S_T-K)^+ \,dK.
$$

Since  
$$g(a) = g'(a) = g''(a) = 0 \quad \text{for all } a \ge U,$$
choosing any $a \ge U$ yields
$$
\mathbb{E}^{\mathbb{Q}}[g(S_T)] 
= \frac{2 e^{rT}}{T} \int_L^U \frac{P(T,K)}{K^2} dK.
$$

---

### Drift term

From part (a),
$$
-\mathbb{E}^{\mathbb{Q}} \left[ \int_0^T g'(S_t) S_t r\,dt \right]
= \mathbb{E}^{\mathbb{Q}}\left[ \int_0^T 
\frac{2r}{T}
\left(
\frac{1}{U}(U - S_t)^+ - \frac{1}{L}(L - S_t)^+
\right) dt \right].
$$

Thus,
$$
-\mathbb{E}^{\mathbb{Q}}\left[\int_0^T g'(S_t) S_t r \, dt \right]
= \frac{2r e^{rT}}{T} \int_0^T 
\left(
\frac{1}{U} P(t, U) - \frac{1}{L} P(t, L)
\right) dt.
$$

---

### Combine the results

$$
\mathbb{E}^{\mathbb{Q}}\left[ \frac{1}{T} \int_0^T 
1_{(L,U)}(S_t)v_t dt \right]
=
\frac{2 e^{rT}}{T} \int_L^U \frac{P(T,K)}{K^2} dK
+
\frac{2r e^{rT}}{T} \int_0^T 
\left( \frac{1}{U}P(t,U) - \frac{1}{L}P(t,L) \right) dt
- g(S_0).
$$

---

### Finally, solve for the break-even strike \(H\)

Setting  
$$
\mathbb{E}^{\mathbb{Q}}[X] 
= \mathbb{E}^{\mathbb{Q}} \left[\frac{1}{T} \int_0^T 1_{(L,U)}(S_t)v_t dt - H \right] = 0
$$  
we obtain

$$
\boxed{
H = 
\frac{2}{T} \left[
e^{rT} \int_L^U \frac{P(T,K)}{K^2} \, dK
+ r e^{rT} \int_0^T \left( \frac{1}{U} P(t,U) - \frac{1}{L} P(t,L) \right) dt
\right] - g(S_0).
}
$$

This expresses the **break-even corridor variance strike** entirely in terms of European put prices.
